```python
import os
```

```python
ImageNamePath=os.listdir('Sences')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('Sences/'+ImageNamePath[0],0)
img=cv2.GaussianBlur(img,(11,11),0)
img=cv2.Canny(img,50,90)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
testCasesRGB=[]
testCasesGray=[]
for i in ImageNamePath:
    try:
        img=cv2.imread('Sences/'+i,1)
        #img=cv2.resize(img,(1000,1000))
        testCasesRGB.append(img)
        img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
        testCasesGray.append(img)
    except:
        pass
```

```python
filterSize=5
T1=10
T2=200
i=5

plt.imshow(testCasesRGB[i])
plt.show()

plt.imshow(testCasesGray[i],cmap='gray')
plt.show()

img=cv2.GaussianBlur(testCasesGray[i],(filterSize,filterSize),0)

img=cv2.Canny(testCasesGray[i],T1,T2)

plt.imshow(img,cmap='gray')
plt.show()

print(testCasesGray[i].shape)
print('---------------------------------------------')
```

```python
image=testCasesGray[0]
plt.imshow(image,cmap='gray')
plt.show()

laplacian=cv2.Laplacian(image,cv2.CV_64F)
plt.imshow(laplacian,cmap='gray')
plt.show()

xSobel=cv2.Sobel(image,cv2.CV_8U,1,0,ksize=5)
plt.imshow(xSobel,cmap='gray')
plt.show()

ySobel=cv2.Sobel(image,cv2.CV_8U,0,1,ksize=5)
plt.imshow(ySobel,cmap='gray')

plt.show()
```

```python
r,g,b=cv2.split(testCasesRGB[0])
```

```python
plt.imshow(r,cmap='gray')
plt.show()
plt.imshow(g,cmap='gray')
plt.show()
plt.imshow(b,cmap='gray')
plt.show()
```

```python
hsv=cv2.cvtColor(testCasesRGB[0],cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(hsv)
cv2.imshow('image',hsv)
cv2.imshow('image',h)
cv2.imshow('image',s)
cv2.imshow('image',v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
yuv=cv2.cvtColor(testCasesRGB[0],cv2.COLOR_BGR2YUV)
y,u,v=cv2.split(yuv)
cv2.imshow('yuv',yuv)
cv2.imshow('yuv',y)
cv2.imshow('yuv',u)
cv2.imshow('yuv',v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
